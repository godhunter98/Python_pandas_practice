A facetgrid class is useful when u want to visualize the distribution of a variable or relationships between variables within subset of our dataset.

```python
g = sns.FacetGrid(data,hue='sex',aspect=4)
g.map(sns.kdeplot,'trestbps',shade=True)
plt.legend()
```
